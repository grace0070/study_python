```python
# data 옆으로 merge 하기 

# 옆으로 통합하기 pd.merge / 아래로 통합 pd.append


#국적코드표 엑셀 파일 불러오기 
code_master = pd.read_excel('/Users/kim-eunhye/Library/CloudStorage/OneDrive-hansung.ac.kr/datasalon-master/02_개정판/2_Data_Analysis_Basic/files/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left = sample_1,
                         right = code_master,
                         how='left',
                         left_on='국적코드',
                         right_on='국적코드')
sample_1_code
```

```python
sample_1_code = pd.merge(left= sample_1,
                         right=code_master,
                         how='inner',  # 두 컬럼값이 일치하는 경우에만 데이터 통합하는 옵션 
                         left_on='국적코드',
                         right_on='국적코드')

sample_1_code
```

```python
sample_2= pd.read_excel('/Users/kim-eunhye/Library/CloudStorage/OneDrive-hansung.ac.kr/datasalon-master/02_개정판/2_Data_Analysis_Basic/files/sample_2.xlsx',
                       header=1,
                       skipfooter=2,
                       usecols='A:C')
sample_2['기준년월'] = '2019-12'
sample_2_code = pd.merge(left=sample_2,
                         right=code_master,
                         how='left',
                         left_on='국적코드',
                         right_on='국적코드')
sample_2_code
```

```python
import pandas as pd
```

```python
sample = pd.concat([sample_1_code, sample_2_code], ignore_index=True)
sample
```

```python
# data save

sample.to_excel('./Users/sample.xlsx', index=False)
```

```python
# 데이터 집계 pivot_table

sample_pivot = sample.pivot_table(values = '입국객수',
                                  index='국적명',
                                  columns='기준년월',
                                  aggfunc='mean')

sample_pivot
```
